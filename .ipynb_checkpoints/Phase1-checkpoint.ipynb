## Phase 1 Data and Fair Value
Notes:
* TLT is an ETF that tracks long-duration US treasuries (20+ yrs)
    * Highly liquid, rate-sensitive (1% rate move $ \approx $ 17% price move)

```python
import pandas as pd
from pathlib import Path
import re

# 1. Define the directory containing your files
data_dir = Path("data/options")

# List to store each processed DataFrame
df_list = []

# 2. Iterate through all CSVs in the target folder
for file_path in data_dir.glob("*.csv"):
    
    # 3. Extract the expiration date using regex
    # Looks for 'exp-' followed by exactly 4 digits, a dash, 2 digits, a dash, 2 digits
    match = re.search(r'exp-(\d{4}-\d{2}-\d{2})', file_path.name)
    
    if match:
        expiration_date = match.group(1)
    else:
        print(f"Warning: Could not find expiration date in {file_path.name}. Skipping.")
        continue
        
    # 4. Read the CSV and trim the bottom row
    df = pd.read_csv(file_path)
    df = df.iloc[:-1].copy() # .copy() prevents SettingWithCopyWarnings later
    
    # 5. Insert the date column (converting to a datetime object for easier time-series analysis later)
    df['Expiration'] = pd.to_datetime(expiration_date)
    
    # Append the clean dataframe to our list
    df_list.append(df)

# 6. Concatenate all DataFrames into one massive table
if df_list:
    final_df = pd.concat(df_list, ignore_index=True)
    print(f"Successfully combined {len(df_list)} files.")
else:
    print("No files were processed.")
    final_df = pd.DataFrame()
```

```python
import yfinance as yf
import requests
import time

# 1. Keep the disguised session to avoid rate limits
session = requests.Session()
session.headers.update({
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
})

max_retries = 3
tlt_data = None

# 2. Build a retry loop
for attempt in range(max_retries):
    try:
        print(f"Attempt {attempt + 1} to fetch TLT...")
        
        # 3. Pass the session AND increase the timeout from 10 to 30 seconds
        tlt_data = yf.download("TLT", period="1y", session=session, timeout=30)
        
        # If the dataframe isn't empty, we got the data! Break the loop.
        if not tlt_data.empty:
            print("Download successful!\n")
            break
            
    except Exception as e:
        print(f"Attempt {attempt + 1} failed with error: {e}")
        
        # If we haven't hit our max retries yet, sleep and try again
        if attempt < max_retries - 1:
            print("Sleeping for 5 seconds before retrying...\n")
            time.sleep(5)
        else:
            print("Max retries reached. Yahoo Finance is unresponsive.")

# Display the data if we successfully grabbed it
if tlt_data is not None and not tlt_data.empty:
    print(tlt_data.head())
```

```python
import yfinance as yf
import pandas as pd
import time

tlt = yf.Ticker("TLT")

# 1. Save all expiry chains
all_chains = []
for expiry in tlt.options[:8]:  # first 8 expiries
    time.sleep(1)               # be polite to the API
    chain = tlt.option_chain(expiry)
    calls = chain.calls.copy()
    puts  = chain.puts.copy()
    calls['type'] = 'call'
    puts['type']  = 'put'
    calls['expiry'] = expiry
    puts['expiry']  = expiry
    all_chains.append(pd.concat([calls, puts]))

options_df = pd.concat(all_chains, ignore_index=True)

# 2. Filter to liquid strikes only
options_df = options_df[options_df['openInterest'] > 200]

# 3. Save immediately — never pull again
options_df.to_csv("tlt_options_chain.csv", index=False)
```
